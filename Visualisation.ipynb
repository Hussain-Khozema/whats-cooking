```python
from collections import Counter
def unique_ingredients(cuisine):
    other_cuisines = []
    for i in cuisine_clean[cuisine_clean.cuisine!= cuisine]['ingredients']:
        for ingredient in i:
            other_cuisines.append(ingredient)
            
    cuisine_unique = []
    for i in all_ingredients:
        if i not in other_cuisines:
            cuisine_unique.append(i)
    counter = Counter()
    for ingredient in cuisine_unique:
        counter[ingredient] += 1
    unique_cuisine = pd.DataFrame(counter.most_common(50), columns = ['ingredients','count'])
    return unique_cuisine
mexican = unique_ingredients('mexican')
mexican.head()
```
